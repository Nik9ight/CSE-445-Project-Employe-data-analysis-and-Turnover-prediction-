#Feature Engeneering

```python
def add_engineered_ratios(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    def safe_div(numer, denom):
        denom_safe = np.where(denom == 0, np.nan, denom)
        res = numer / denom_safe
        return np.nan_to_num(res, nan=0.0, posinf=0.0, neginf=0.0)

    required = [
        "YearsAtCompany", "TotalWorkingYears", "MonthlyIncome",
        "YearsInCurrentRole", "YearsSinceLastPromotion", "TrainingTimesLastYear"
    ]
    missing = [c for c in required if c not in out.columns]
    if missing:
        raise ValueError(f"Missing required columns for feature engineering: {missing}")

    # 1) Loyalty ratio
    out["loyalty_ratio"] = safe_div(out["YearsAtCompany"], out["TotalWorkingYears"])
    # 2) Income growth proxy
    out["income_growth_proxy"] = safe_div(out["MonthlyIncome"], out["TotalWorkingYears"])
    # 3) Promotion speed
    out["promotion_speed"] = safe_div(out["YearsInCurrentRole"], out["YearsAtCompany"])
    # 4) Career stagnation
    out["career_stagnation"] = safe_div(out["YearsSinceLastPromotion"], out["YearsAtCompany"])
    # 5) Training intensity
    out["training_intensity"] = safe_div(out["TrainingTimesLastYear"], out["TotalWorkingYears"])

    return out
```

# Function for model train and evalution

```python
def evaluate_model(model, X_tr, y_tr, X_v, y_v, X_te, y_te, name, feature_names):
    model.fit(X_tr, y_tr)

    # Validation
    y_val_pred = model.predict(X_v)
    print(f"\n=== {name} - Validation Performance ===")
    print("Accuracy:", accuracy_score(y_v, y_val_pred))
    print("Precision:", precision_score(y_v, y_val_pred))
    print("Recall:", recall_score(y_v, y_val_pred))
    print("F1 Score:", f1_score(y_v, y_val_pred))
    print("ROC-AUC:", roc_auc_score(y_v, model.predict_proba(X_v)[:, 1]))
    print(classification_report(y_v, y_val_pred))

    # Test
    y_test_pred = model.predict(X_te)
    print(f"\n=== {name} - Test Performance ===")
    print("Accuracy:", accuracy_score(y_te, y_test_pred))
    print("Precision:", precision_score(y_te, y_test_pred))
    print("Recall:", recall_score(y_te, y_test_pred))
    print("F1 Score:", f1_score(y_te, y_test_pred))
    print("ROC-AUC:", roc_auc_score(y_te, model.predict_proba(X_te)[:, 1]))
    print(classification_report(y_te, y_test_pred))

    # Feature Importance Plot
    if hasattr(model, "feature_importances_"):
        importances = model.feature_importances_
        indices = np.argsort(importances)[::-1]
        plt.figure(figsize=(10, 6))
        plt.title(f"{name} - Top 15 Feature Importances")
        plt.bar(range(15), importances[indices][:15], align="center")
        plt.xticks(range(15), [feature_names[i] for i in indices[:15]], rotation=90)
        plt.tight_layout()
        plt.show()
```

```python
def eval_on(split_name, model, X, y):
    y_prob = model.predict_proba(X)[:, 1]
    y_hat  = (y_prob >= 0.5).astype(int)
    print(f"\n[{split_name}]",
          "AUC:", roc_auc_score(y, y_prob),
          "Acc:", accuracy_score(y, y_hat),
          "Prec:", precision_score(y, y_hat),
          "Rec:", recall_score(y, y_hat),
          "F1:", f1_score(y, y_hat))
    print(classification_report(y, y_hat))
```

```python
# CV and scorer
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
recall_pos = make_scorer(recall_score, pos_label=1)

xgb_pipe = Pipeline(steps=[
    ("adasyn", ADASYN(random_state=42)),
    ("scaler", StandardScaler(with_mean=True)),
    ("clf", XGBClassifier(
        objective="binary:logistic",
        eval_metric="auc",
        tree_method="hist",
        random_state=42,
        n_jobs=-1
    ))
])

xgb_param_dist = {
    "clf__n_estimators":        np.arange(200, 700, 50),
    "clf__learning_rate":       np.linspace(0.01, 0.2, 10),
    "clf__max_depth":           np.arange(3, 9),
    "clf__min_child_weight":    np.arange(1, 8),
    "clf__subsample":           np.linspace(0.6, 1.0, 5),
    "clf__colsample_bytree":    np.linspace(0.6, 1.0, 5),
    "clf__gamma":               np.linspace(0, 4, 9),
    "clf__reg_lambda":          np.logspace(-2, 2, 7),
}

xgb_search = RandomizedSearchCV(
    estimator=xgb_pipe,
    param_distributions=xgb_param_dist,
    n_iter=60,
    scoring={"roc_auc": "roc_auc", "recall": recall_pos},
    refit="roc_auc",
    cv=cv,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

xgb_search.fit(X_train, y_train)   # ADASYN+scaling happen inside folds
print("Best XGB AUC (CV):", xgb_search.best_score_)
print("Best XGB params:", xgb_search.best_params_)
best_xgb_cv_model = xgb_search.best_estimator_

# Evaluate tuned pipeline (no refit yet) on val + test
eval_on("Validation (XGB tuned pipeline)", best_xgb_cv_model, X_val, y_val)
eval_on("Test (XGB tuned pipeline)",       best_xgb_cv_model, X_test, y_test)
```

```python
rf_pipe = Pipeline(steps=[
    ("adasyn", ADASYN(random_state=42)),
    ("scaler", StandardScaler(with_mean=True)),  # harmless for RF, keeps parity
    ("clf", RandomForestClassifier(
        n_estimators=400,
        random_state=42,
        n_jobs=-1
    ))
])

rf_param_dist = {
    "clf__n_estimators":       np.arange(200, 801, 100),
    "clf__max_depth":          [None] + list(np.arange(4, 31, 2)),
    "clf__min_samples_split":  [2, 5, 10, 20],
    "clf__min_samples_leaf":   [1, 2, 4, 8],
    "clf__max_features":       ["sqrt", "log2", 0.5, 0.7, 1.0],
    # "clf__class_weight":     [None, "balanced"]  # usually not needed with ADASYN
}

rf_search = RandomizedSearchCV(
    estimator=rf_pipe,
    param_distributions=rf_param_dist,
    n_iter=50,
    scoring={"roc_auc": "roc_auc", "recall": recall_pos},
    refit="roc_auc",
    cv=cv,
    verbose=1,
    n_jobs=-1,
    random_state=42
)

rf_search.fit(X_train, y_train)
print("Best RF AUC (CV):", rf_search.best_score_)
print("Best RF params:", rf_search.best_params_)
best_rf_cv_model = rf_search.best_estimator_

eval_on("Validation (RF tuned pipeline)", best_rf_cv_model, X_val, y_val)
eval_on("Test (RF tuned pipeline)",       best_rf_cv_model, X_test, y_test)
```

#Learning curve

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

def plot_learning_curve(estimator, X, y, title):
    train_sizes, train_scores, val_scores = learning_curve(
        estimator, X, y,
        cv=5,
        scoring="roc_auc",
        n_jobs=-1,
        train_sizes=np.linspace(0.1, 1.0, 5),
        random_state=42
    )
    train_mean, val_mean = np.mean(train_scores, axis=1), np.mean(val_scores, axis=1)
    plt.plot(train_sizes, train_mean, "o-", label="Training AUC")
    plt.plot(train_sizes, val_mean, "o-", label="Validation AUC")
    plt.title(title)
    plt.xlabel("Training set size")
    plt.ylabel("ROC-AUC")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_learning_curve(best_xgb_cv_model, X_train, y_train, "XGBoost Learning Curve")
plot_learning_curve(best_rf_cv_model, X_train, y_train, "Random Forest Learning Curve")
plot_learning_curve(best_knn, X_train, y_train, "KNN Learning Curve")
plot_learning_curve(best_logreg, X_train, y_train, "Logistic Regression Learning Curve")

```

#ROC-AUC

```python
from sklearn.metrics import roc_curve, auc

def plot_roc_curve(models, X, y, title):
    plt.figure(figsize=(7,6))
    for name, model in models.items():
        y_prob = model.predict_proba(X)[:,1]
        fpr, tpr, _ = roc_curve(y, y_prob)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")
    plt.plot([0,1], [0,1], "k--", label="Chance")
    plt.title(title)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.legend()
    plt.grid(True)
    plt.show()

models = {
    "XGBoost": best_xgb_cv_model,
    "Random Forest": best_rf_cv_model,
    "KNN": best_knn,
    "Logistic Regression": best_logreg
}

plot_roc_curve(models, X_test, y_test, "ROC Curves (Test Set)")
```

#Pricesion-Recall Curve

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

def plot_pr_curve(models, X, y, title):
    plt.figure(figsize=(7,6))
    for name, model in models.items():
        y_prob = model.predict_proba(X)[:,1]
        prec, rec, _ = precision_recall_curve(y, y_prob)
        ap = average_precision_score(y, y_prob)
        plt.plot(rec, prec, label=f"{name} (AP={ap:.3f})")
    plt.title(title)
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.legend()
    plt.grid(True)
    plt.show()

plot_pr_curve(models, X_test, y_test, "Precision-Recall Curves (Test Set)")
```
